<h2 style="font-size: 25px; font-weight: bold; color: #fb6227;">
  4. Deployment of a Mistral AI Model on Microsoft Azure
</h2>

**Objectives**: Deploy and use a Mistral AI model hosted by Microsoft Azure

**Difficulty**: 🟢 ⚪️ ⚪️ ⚪️ - Only some basic Python knowledge is required.

---
<h3 style="font-size: 20px; font-weight: bold; color: #ff8f1e;">
  4.1 Create an Azure Foundry Resource
</h3>

1. Create the resource on [this page](https://portal.azure.com/#view/Microsoft_Azure_ProjectOxford/CognitiveServicesHub/~/AIFoundryPhaseTwo_overview)
2. Open your resource and deploy a model
3. Copy the API key into your .env file
4. Copy the URL for inclusion in your MistralAzure instance

---

<h3 style="font-size: 20px; font-weight: bold; color: #ff8f1e;">
  4.2 Access Model via the Python SDK
</h3>

- We use `mistralai.azure.client` instead of `mistralai.client`and `MistralAzure` instead of `Mistral`
- SDK models are also included in `mistralai.azure.client.models` instead of just `mistralai.client.models`
- The model parameter corresponds to the Azure deployment name (often the same as the model name but not always!)

```python
from mistralai.azure.client import MistralAzure
from mistralai.azure.client.models import UserMessage
from dotenv import load_dotenv
import os

load_dotenv()

mistral = MistralAzure(
    api_key=os.environ["AZURE"],
    server_url="https://mistral-new-foundry.services.ai.azure.com/models/chat/completions?api-version=2024-05-01-preview"
)

response = mistral.chat.complete(
    model="Ministral-3B", # Use deployment name!
    messages=[UserMessage(content="What is the most popular place to visit in Austria?")]
)

response.choices[0].message.content
```
